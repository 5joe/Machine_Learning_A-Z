# Support Vector Regression

### Importing the dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
df = pd.read_csv("../01_dataset/Data_model.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
y = y.reshape(len(y), 1)
```

```python
print(y)
```

#### Splitting the data into the training and the test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```python
print(X_train)
```

```python
print(y_train)
```

### Training the SVR model on the whole dataset

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
```

### Predicting the new result

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test)).reshape(-1,1))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
sc_y.inverse_transform(regressor.predict(sc_X.transform([[6.5, 4.5, 400, 26]])).reshape(-1,1))

# -- note that the inverse_tranform helps handle the fact that the y values that were used to train the models had been feature scaled too --
```

### Checking the accuracy of the model r2_score

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
