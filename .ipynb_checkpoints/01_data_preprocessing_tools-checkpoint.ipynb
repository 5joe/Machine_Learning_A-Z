# Data Preprocessing Tools

## Importing the libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("01_dataset/Data.csv")
x = dataset.iloc[:, :-1].values  # Rows first, then columns next(-1 is the index for the last column)
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

## Taking Care of Missing Data

```python
## In this case we are replacing the missing data with mean for the numeric columns
# This is the Class
from sklearn.impute import SimpleImputer
# This is the object of the class
imputer = SimpleImputer(missing_values=np.nan, strategy = 'mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

```python
print(x)
```

## Encoding Categorical data

## Encoding the independent variables

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers =[('encoder',OneHotEncoder(), [0])], remainder = 'passthrough')
X = np.array(ct.fit_transform(x))
```

```python
print(X)
```

## Encoding the Dependent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## Splitting the dataset into the Training set and the Test set

```python
# We do splitting before Feature scaling to prevent information leakage on the Test set which we are supposed to have until the trainig is done.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
print(y_train)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# Do not apply Feature Scaling on the dummy variables that are mapped to the Countries column as
#this will cause you to lose the meaning and country mapping
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
# remember you need to use the same scaler as that was used on the X_train
#therefore do not get another fit_scaler i.e fit_transform
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```python
print(X_train)
```

```python
print(X_test)
```
